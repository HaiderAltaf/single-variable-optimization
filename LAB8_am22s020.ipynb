##  Plot the surface J(w1, w2) = (w1 - 10)^2 + (w2 - 10)^2. Also, generated the corresponding contour plot

```python
import numpy as np
import matplotlib.pyplot as plt

w1=np.arange(-500,500,1)
w2=np.arange(-500,500,1)

X,Y = np.meshgrid(w1, w2)  # create 2D grid

J=(X-10)**2 +(Y-10)**2

# surface plot


fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, J, 50, cmap='binary')
ax.set_title('Surface Plot')
ax.set_xlabel('w1')
ax.set_ylabel('w2')
ax.set_zlabel('J');
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

w1=np.arange(-500,500,1)
w2=np.arange(-500,500,1)

# contour plot

[X, Y] = np.meshgrid(w1, w2)

J=(X-10)**2 +(Y-10)**2

fig, ax = plt.subplots(1,1) 

ax.contour(X, Y, J)
  
ax.set_title('Contour Plot')
ax.set_xlabel('w1')
ax.set_ylabel('w2')
  

cp=plt.contour(X,Y,J)
plt.clabel(cp, fontsize=8)
plt.show()
```

```python
#  (a) bracketing method (choose you own a, b, n)

# Minimize f(x) = w2 + 54/w
# using exhaustive search method

def func(w):
    return w**2 + 54/w


a=int(input("enter lower limit of interval: "))
b=int(input("enter upper limit of interval: "))
n=int(input("enter number of intermediate points: "))

delta_w=(b-a)/n

w1=a
w2=w1+delta_w
w3=w2+delta_w

for i in range(n):
    f1=func(w1)
    f2=func(w2)
    f3=func(w3)
    
    if f1>f2>f3 and w3<b:
        w1=w2
        w2=w3
        w3=w3+delta_w
        
    elif f1>f2<f3:
        print(f"Mininum lies in range ({w2},{w3})")
        break
    else:
        print(f"No minima in the range of {a} to {b} of value of x")
        break
```
